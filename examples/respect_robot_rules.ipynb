```python
# -- Import package --
from imagescraper import Browser, Location, Scraper
```

```python
# -- Create and launch browser --
browser = Browser({'disrespect_robot_rules': False})
```

```python
# -- Define search query --
query = 'vienna'
```

```python
# -- Create location to fetch --
location = Location(
    hostname='www.google.com',
    pathname='search',
    params={
        'q': query,
        'source': 'lnms',
        'tbm': 'isch',
    },
)
```

```python
# -- Create scraper --
# Scraper takes location to scrap and browser to scrap with
scraper = Scraper(
    location,
    browser,
    {
        'img_min_height': 100,
        'max_img_count': 100,
        'save_dir': 'images/{}'.format(query),
        'sleep_time': 0.3,
    }
)
```

```python
# Start scraping
scraper.run()
```
