```python
# -- Import package --
from imagescraper import Browser, Location, Scraper
```

```python
# -- Create and launch browser --
browser = Browser()
```

```python
# -- Define search query --
query = 'vienna'
```

```python
# -- Create location to fetch --
location = Location(
    hostname='www.shutterstock.com',
    pathname='search/{query}'.format(query=query),
)
```

```python
# -- Create scraper --
# Scraper takes location to scrap and browser to scrap with
scraper = Scraper(
    location,
    browser,
    {
        'img_min_height': 100,
        'max_img_count': 100,
        'save_dir': 'images/{}'.format(query),
        'sleep_time': 0.3,
    }
)
```

```python
# Start scraping
scraper.run()
```
